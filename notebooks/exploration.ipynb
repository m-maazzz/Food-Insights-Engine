```python
import pandas as pd
```

```python
def inspect_df(df, name):
    """
    Display useful information about a dataframe
    """

    print("=" * 60)
    print(f"DATAFRAME: {name}")
    print("=" * 60)

    print("\nShape:")
    print(df.shape)

    print("\nColumns:")
    print(df.columns.tolist())

    print("\nData Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nFirst 5 Rows:")
    print(df.head())

    print("\n")
```
